```python
import textblob
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/spam.csv',sep=',',encoding='ISO-8859-1')
df.head()
```

```python
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1,inplace=True)
df.info()
```

```python
df.columns=['label','sms']
df.head()
```

```python
import string
string.punctuation

def remove_punctuation(txt):
    txt_nopunct = "".join([c for c in txt if c not in string.punctuation])
    return txt_nopunct
```

```python
df['msg_clean'] = df['sms'].apply(lambda x: remove_punctuation(x))
df.head()
```

```python
from textblob import TextBlob

def correct_spelling(sentences):
  sentence = TextBlob(sentences)
  result = sentence.correct()
  return result

df['corrected'] = df['msg_clean'].apply(lambda x: correct_spelling(x))
```
